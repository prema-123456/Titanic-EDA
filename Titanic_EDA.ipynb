Import Required Libraries

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


plt.style.use('default')
````

Load the Dataset

````python
df = pd.read_csv('Titanic.csv')
df.head()
````

Basic Dataset Inspection

````python
df.shape
df.info()
df.describe()
````

Check Missing Values

````python
df.isnull().sum()
````

### Missing Value Check
- The dataset contains no missing values.
- No imputation or row/column removal was required.
- This ensures data integrity is preserved for further analysis.



Univariate Analysis (Single Variable)

Survival Count

````python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
````

Gender Distribution

````python
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()
````

Bivariate Analysis (Relationship with Survival)

Survival by Gender

````python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
````

 Insight: Females had a higher survival rate.

Survival by Passenger Class

````python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
````

 Insight: 1st class passengers survived more.

Age vs Survival

````python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survival')
plt.show()
````

Fare Analysis

````python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare vs Survival')
plt.show()
````

 Insight: Higher fare  higher survival probability.

Correlation Heatmap

````python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
````

Feature Engineering

Create Age Groups:

````python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0,12,18,35,60,80],
                        labels=['Child','Teen','Young Adult','Adult','Senior'])
````

````python
sns.countplot(x='AgeGroup', hue='Survived', data=df)
plt.xticks(rotation=45)
plt.show()
````

Final Insights

### Key Insights:
- Female passengers had significantly higher survival rates.
- First-class passengers were more likely to survive.
- Passengers who paid higher fares had better survival chances.
- Children showed relatively higher survival compared to adults.
